```python
import pyarrow.parquet as pq
import pandas as pd

table = pq.read_table("sikhna.parquet")
df = table.to_pandas()

print(df.head())
```

```python
from datasets import load_dataset

dataset = load_dataset("parquet", data_files="sikhna.parquet")
print(dataset)
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os

# Find a Devanagari-capable font on the system
_devanagari_font = None
_candidates = [
    # Windows — Nirmala UI supports full Devanagari
    'C:/Windows/Fonts/Nirmala.ttc',
    'C:/Windows/Fonts/Nirmala.ttf',
    'C:/Windows/Fonts/NirmalaB.ttf',
    # Noto fonts if installed
    'C:/Windows/Fonts/NotoSansDevanagari-Regular.ttf',
]
for _path in _candidates:
    if os.path.exists(_path):
        _devanagari_font = fm.FontProperties(fname=_path)
        break

def show_hindi_image(image, text, ax=None):
    if ax is None:
        fig, ax = plt.subplots()
    ax.imshow(image)
    ax.axis("off")
    if _devanagari_font:
        ax.set_title(text, fontproperties=_devanagari_font, fontsize=14)
    else:
        ax.set_title(text, fontsize=14)
    return ax

import matplotlib.pyplot as plt
from PIL import Image
import io

for i in range(10):
    row = df.iloc[i]

    image_bytes = row["image"]["bytes"]
    text = row["text"]

    image = Image.open(io.BytesIO(image_bytes))

    fig, ax = plt.subplots(figsize=(6, 3))
    show_hindi_image(image, text, ax)
    plt.tight_layout()
    plt.show()
```
